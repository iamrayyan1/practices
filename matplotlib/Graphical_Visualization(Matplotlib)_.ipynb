#Side – by – Side Bargraphs:

```python
import numpy as np
# Data
gender = ['Male', 'Female', 'Other']
ratio = [10, 15, 7]
ratio2 = [2, 5, 7]

# Colors
defined_colors = ['r', 'b', 'g']
defined_colors2 = ['y', 'orange', 'pink']

# Bar width
bar_width = 0.3

# X positions for each group of bars
x = np.arange(len(gender))  # Positions for the first set of bars
x2 = x + bar_width  # Offset the second set of bars

# Create the bars
plt.bar(x, ratio, width=bar_width, color=defined_colors, label='Ratio 1')
plt.bar(x2, ratio2, width=bar_width, color=defined_colors2, label='Ratio 2')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Ratio')
plt.title('Side-by-Side Bar Graph')

# Add x-ticks to the middle of both sets of bars
plt.xticks(x + bar_width / 2, gender)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

#** =============Scatter Plot===============

```python
# Data for days and expenses
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week or any other time frame
expenses = [200, 300, 250, 400, 150, 350, 300]  # Expenses on each day

# Create a scatter plot
plt.scatter(days, expenses, color='b', marker='o')

# Add labels and title
plt.xlabel('Days')
plt.ylabel('Expenses (in $)')
plt.title('Scatter Plot of Days vs. Expenses')

# Show the plot
plt.show()
```

```python
# Colors & sizes:
import matplotlib.pyplot as plt

# Data for days and expenses
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week or any other time frame
expenses = [200, 300, 250, 400, 150, 350, 300]  # Expenses on each day

# Define colors for each data point
colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange', 'pink']

# Define sizes for each data point
sizes = [100, 200, 150, 300, 50, 250, 180]  # Size of each point

# Create a scatter plot with varying colors and sizes
plt.scatter(days, expenses, c=colors, s=sizes, alpha=0.7, edgecolor='black')

# Add labels and title
plt.xlabel('Days')
plt.ylabel('Expenses (in $)')
plt.title('Scatter Plot of Days vs. Expenses with Custom Colors and Sizes')

# Show the plot
plt.show()
```

```python
#marker
# Create a scatter plot with varying colors, sizes, and a custom marker
plt.scatter(days, expenses, c=colors, s=sizes, marker='D', alpha=0.7, edgecolor='black')
```

#Histogram

```python
# Example data (expenses on different days)
expenses = [200, 300, 250, 400, 150, 350, 300, 220, 270, 310, 180, 400, 150, 340, 370]

# Create the histogram
plt.hist(expenses,width=20, bins=5, color='blue', edgecolor='black')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data (expenses on different days)
expenses = [200, 300, 250, 400, 150, 350, 300, 220, 270, 310, 180, 400, 150, 340, 370]

# Create the cumulative histogram
plt.hist(expenses, bins=5, color='blue', edgecolor='black', cumulative=True)

# Add labels and title
plt.xlabel('Expenses (in $)')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Histogram of Expenses')

# Show the plot
plt.show()
```

#Setting type in a histogram:

The histtype parameter in matplotlib's plt.hist() function controls the style or type of the histogram. It allows you to choose how the bars in the histogram are displayed. The available histtype options are:

'bar' (default): Standard bar histogram where each bin is represented by a bar.
'barstacked': Multiple data sets are stacked on top of each other.
'step': A line plot representing the histogram without filled bars (just the edges or steps).
'stepfilled': A filled version of the step histogram.

```python
# Create the histogram with the default 'bar' type
plt.hist(expenses, bins=5, color='blue', edgecolor='black', histtype='bar')
```

```python
# Two example datasets (expenses for two weeks)
expenses_week1 = [200, 300, 250, 400, 150]
expenses_week2 = [350, 300, 220, 270, 310]

plt.hist([expenses_week1, expenses_week2], bins=5, color=['blue', 'green'], edgecolor='black', histtype='barstacked')
```

```python
# Create the step histogram
plt.hist(expenses, bins=5, color='blue', edgecolor='black', histtype='step')
```

```python
# Create the step-filled histogram
plt.hist(expenses, bins=5, color='blue', edgecolor='black', histtype='stepfilled')
```

#Changing orientation in a histogram:

```python
# Create the step histogram
plt.hist(expenses, bins=5, color='blue', edgecolor='black',orientation='horizontal')
```

#**=========== Pie Charts ===============**

```python
gender = ['Male', 'Female', 'Other']
ratio = [10, 15, 7]

# Create a pie chart
plt.pie(ratio, labels=gender, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.show()
```
