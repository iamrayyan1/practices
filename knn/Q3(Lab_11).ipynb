Q3: Download the Dermatology data set from the UCI depositary, apply KNN and analyze the
performance. Display the confusion matrix and discuss it. Use 10-Fold Cross Validation and random train/test split (70%, 30%).
Note: There might be some missing/null values in the dataset. First clean the data and then
work on the algorithm.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('/content/dermatology.data')

df.replace('?', np.nan, inplace=True)
df.head()
```

```python
print(df.isnull().sum())

# to impute missing values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df = pd.DataFrame(imputer.fit_transform(df))

print(df.isnull().sum())
```

```python
X = df.iloc[:, :-1]  # All columns except the last one
y = df.iloc[:, -1]   # The last column as target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix\n", cm)

print("Classification Report: \n",classification_report(y_test, y_pred))
```

10-Fold Cross-Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn, X, y, cv=10)
print("Cross validation scores:", scores)
print("Mean accuracy:", scores.mean())
```
